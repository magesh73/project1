## Edge Detection in image processing.

**What is Edge ?**

In image processing edge is an abrupt change in intensity of pixels or discontinuity in the brightness or contrast of pixels compared to its neighboring cells.

**Why we need to detect Edge ?**

Edge detection helps in reducing noise, extract salient features like lines, curves, and corners, and recognizes objects, boundaries and aids in segmentation of image.


**What are method’s available for edge detection?**

•	Gradient(1st order derivative)

•	Zero crossing methods (second order derivative) 

•	Gaussian

•	Fuzzy logic

We will be looking at Gradient(1st order derivative) edge detectors, Gradient based detection uses operators like ROBERT, PREWITT, SOBEL,CANNY to detect edges, pls see below sample code that demonstrate different Edge detection operators.

**Advantage’s of Gradient based detection**

•	Best used in abrupt discontinuities

•	Perform better in less noised image

**Disadvantage of Gradient based detection**

•	NOT accurate in LOCAtING EDGES

```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

**Prewit Operator**



Prewitt edge detector measures gradient vector for a Pixel(x,y) by calculating partial derivative on x direction(**horizontal**) which is color difference in neighboring cells that is on left and right and on y direction(**vertical**) the color difference between neighboring pixels above and below.

To detect Horizonal  and vertical edge, two 3x3 filter is applied, though below sample shows it sepetately for illustation purpose, they can still paired togther to identify edges

![alt text](https://www.researchgate.net/profile/S_N_Kumar/publication/317754223/figure/fig3/AS:565335482351616@1511797890242/Masks-for-the-Prewitt-gradient-edge-detector-The-Laplacian-operator-is-based-on-second.png)


```
#Prewitt HORIZONTAL detector
kernel = np.float32([[1,1,1],[0,0,0],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Horizontal')
plt.xticks([]), plt.yticks([])
plt.show()
```

```
#Prewitt VERTICAL detector
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Vertical')
plt.xticks([]), plt.yticks([])
plt.show()
```

**45 Degree Angle Detector**
 
Roberts Edge detector measures** gradient vector** for a Pixel(x,y) by calculating partial derivative on x direction(horizontal) which is color difference in neighboring cells that is on **left** and **right** and on y direction(vertical) the color difference between neighboring pixels **above** and **below**.


Roberts Edge detector  masks can be applied to the input image, to produce separate measurements of the gradient component
in each orientation (call these Gx and Gy). These can then be combined together to find the
absolute magnitude of the gradient at each point and the orientation of that gradient. The

To detect 45 degree edges  3x3  filter is used.


```
#Roberts 45 Degree Angle Detector 3X3 Filter
kernel = np.float32([[0,1,2],[0,-1,0],[-2,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('45 Degree angle filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

Blur Kernel

The blur kernel de- emphasizes differences in adjacent pixel values. 


```
#Blur detector 
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(image,-1,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Blur')
plt.xticks([]), plt.yticks([])
plt.show()
```

Sharpen Kernel


The sharpen kernel emphasizes differences in adjacent pixel values. This makes the image look more vivid.
